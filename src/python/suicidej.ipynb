```python
# ライブラリをインポート

# データ操作と数値計算のため
import pandas as pd
import numpy as np

# データ可視化のため
import matplotlib.pyplot as plt
import seaborn as sns

# 統計モデリングと計量経済分析のため
import statsmodels.api as sm
from statsmodels.iolib.summary2 import summary_col

# for ignore warning
import warnings
warnings.filterwarnings('ignore')
```

```python
url = "https://www.fbc.keio.ac.jp/~tyabu/keiryo/suicidej_data.csv"
df = pd.read_csv(url)
df.head()
```

```python
model_1 = sm.OLS.from_formula('suicide ~ unemployment + C(prefecture)', data=df)
results_1 = model_1.fit(cov_type='cluster', cov_kwds={'groups': df['prefecture']})
print(results_1.summary())
```

```python
model_2 = sm.OLS.from_formula('suicide ~ unemployment + C(prefecture) + C(year)', data=df)
results_2 = model_2.fit(cov_type='cluster', cov_kwds={'groups': df['prefecture']})
print(results_2.summary())
```
